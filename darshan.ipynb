```python
from PIL import Image
import numpy as np
```

```python
img = Image.open("darshan.jpg").convert("RGB")
```

```python
WIDTH = 120
aspect_ratio = img.height / img.width
HEIGHT = int(WIDTH * aspect_ratio * 0.55)

img = img.resize((WIDTH, HEIGHT))
```

```python
pixels = np.array(img)
```

```python
ASCII_CHARS = " .:-=+*#%@"
```

```python
def get_ascii_char(brightness):
    if brightness < 50:
        return ASCII_CHARS[8]
    elif brightness < 80:
        return ASCII_CHARS[7]
    elif brightness < 110:
        return ASCII_CHARS[6]
    elif brightness < 140:
        return ASCII_CHARS[5]
    elif brightness < 170:
        return ASCII_CHARS[4]
    elif brightness < 200:
        return ASCII_CHARS[3]
    elif brightness < 220:
        return ASCII_CHARS[2]
    else:
        return ASCII_CHARS[1]
```

```python
def color_pixel(char, r, g, b):
    return f"\033[38;2;{r};{g};{b}m{char}"
```

```python
ascii_image = ""
```

```python
for y in range(HEIGHT):
    for x in range(WIDTH):
        r, g, b = pixels[y][x]

        brightness = (0.299*r + 0.587*g + 0.114*b)

        # slight overall dark
        brightness *= 0.88

        # TEETH WHITENING (mouth region)
        if (
            brightness > 200 and
            abs(r - g) < 12 and abs(g - b) < 12 and
            HEIGHT * 0.45 < y < HEIGHT * 0.65
        ):
            r, g, b = 255, 255, 255

        char = get_ascii_char(brightness)
        ascii_image += color_pixel(char, r, g, b)

    ascii_image += "\n"
```

```python
print(ascii_image)
```
